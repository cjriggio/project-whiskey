```python
title = soup2.find_all('div', {'class': 'div-block-35'})
body = soup2.find_all('div', {'class': 'desktoptext w-richtext'})

keys = []
values = []

for i in range(len(title)-1):
    titles = title[i].text
    descrip = body[i+1].text
    
    keys.append(titles)
    values.append(descrip)
    
    bottle_dic.update({keys[i]:values[i]})
```
