```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
def scrape(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    return soup
```

```python
bottle = scrape('https://www.breakingbourbon.com/review/makers-46-bourbon')
```

```python
bottle_info = bottle.find_all('div', {'class': 'bottleinfo w-richtext'})[0]
bottle_info2 = bottle_info.find_all('p')

bt_info_list = []
for i in range(len(bottle_info2)-1):
    text = bottle_info2[i].text
    bt_info_list.append(text)

keys1 = []
values1 = []

for i in range(len(bt_info_list)-1):
    splt = bt_info_list[i].split(':')
    keys1.append(splt[0])
    values1.append(splt[1])
    
bottle_dic = {keys1[i]: values1[i][1:] for i in range(len(keys1))} 

title = bottle.find_all('div', {'class': 'div-block-35'})
body = bottle.find_all('div', {'class': 'desktoptext w-richtext'})

keys2 = []
values2 = []

for i in range(len(title)-1):
    titles = title[i].text
    descrip = body[i+1].text
    
    keys2.append(titles)
    values2.append(descrip)
    
    bottle_dic.update({keys2[i]:values2[i]})
```
