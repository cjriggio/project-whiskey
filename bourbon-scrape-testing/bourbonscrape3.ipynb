```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import unicodedata
```

```python
def scrape(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    return soup
```

```python
bourbon_names = scrape('https://www.breakingbourbon.com/bourbon-rye-whiskey-reviews')

names1 = bourbon_names.find_all('div', attrs={'class': 'collection-item-11 w-dyn-item'})
name_list = []

for i in range(len(names1)):
    names2 = names1[i]
    names3 = names2.find('a').text
    name_list.append(names3)

url_list = []
url1 = bourbon_names.find_all('div', {'class':'collection-item-11 w-dyn-item'})

for i in range(len(name_list)):
    url2 = url1[i].find_all('a')[0]
    url_list.append('https://www.breakingbourbon.com' + url2.get('href'))
```

```python
final2 = []

for y in range(128):
    bttle = scrape(url_list[y+115])
    bottle_info = bttle.find_all('div', {'class': 'bottleinfo w-richtext'})[0]
    bottle_info2 = bottle_info.find_all('p')



    bt_info_list = []

    for g in range(len(bottle_info2)-1):
        info = unicodedata.normalize("NFKD", bottle_info2[g].text)
        bt_info_list.append(info)

    keys1 = []
    values1 = []

    for i in range(len(bt_info_list[:6])):
        splt = bt_info_list[i].split(':')
        keys1.append(splt[0])
        values1.append(splt[1])

    bottle_dic = {keys1[i]: values1[i][1:] for i in range(len(keys1))}

    keys2 = []
    values2 = []

    for j in range(len(bt_info_list[-2:])):
        splt2 = bt_info_list[0-j].split(':')
        keys2.append(splt2[0])
        values2.append(splt2[1])

        bottle_dic.update({keys2[j]:values2[j]})

    title = bttle.find_all('div', {'class': 'div-block-35'})
    body = bttle.find_all('div', {'class': 'desktoptext w-richtext'})

    keys3 = []
    values3 = []

    for k in range(len(title)-1):

        titles = title[k].text
        descrip = body[k+1].text

        keys3.append(titles)
        values3.append(descrip)

        bottle_dic.update({keys3[k]:values3[k]})

    final2.append(bottle_dic)
    print(y)
```
